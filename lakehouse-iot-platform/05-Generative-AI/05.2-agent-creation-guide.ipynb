### Make sure you add a proper system prompt so that your agent knows how to use the AI function you provided.

Here is an example (‚úÖ **Free Edition Compatible** - uses only 2 tools):

```
Act as an assistant for wind turbine maintenance technicians.
    These are the tools you can use to answer questions:
      - turbine_specifications_retriever: takes turbine_id as input and retrieves the sensor_readings.
      - turbine_maintenance_predictor: takes as input sensor_readings and predicts whether or not a turbine is at risk of failure.
        Use turbine_specifications_retriever to get the current status.

If a user gives you a turbine ID, first look up that turbine's information with turbine_specifications_retriever, then use turbine_maintenance_predictor to check if the turbine is at risk of failure.
```

‚ö†Ô∏è **Note:** The original demo included a 3rd tool (`turbine_maintenance_guide_retriever`) that uses Vector Search. Vector Search is not available in Databricks Free Edition, so we focus on the two tools above.

### You can now ask questions such as:

* Get me the status of the wind turbine ID '004a641f-e9e5-9fff-d421-1bf88319420b'. Is it at risk of failure?
* What are the current sensor readings for turbine '004a641f-e9e5-9fff-d421-1bf88319420b'?
* Check if any turbines need maintenance based on their sensor data.
